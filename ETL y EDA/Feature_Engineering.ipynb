# Feature Engineering

```python
import pandas as pd
from textblob import TextBlob
```

Cargamos el dataset limpio de australian_user_reviews a un Dataframe

```python
user_reviews = pd.read_csv('australian_user_reviews.csv')
```

Definimos una función para realizar el análisis de sentimiento

```python
def sent_analysis(review):
    if isinstance(review, str):
        analysis = TextBlob(review)
        if analysis.sentiment.polarity < 0:
            return 0  # Sentimiento negativo
        elif analysis.sentiment.polarity == 0:
            return 1  # Sentimiento neutral
        else:
            return 2  # Sentimiento positivo
    else:
        return 1 #Si no hay reseña, se considera neutral
```

Agregamos la columna solicitada

```python
user_reviews['sentiment_analysis'] = user_reviews['review'].apply(sent_analysis)
```

Mostramos el Dataframe actualizado

```python
user_reviews
```

Guardamos el Dataframe actualizado en formato CSV

```python
user_reviews.to_csv('australian_user_reviews_actualizado.csv', index=False)
```
