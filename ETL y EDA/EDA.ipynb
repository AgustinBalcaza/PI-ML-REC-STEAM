Podemos comenzar a identificar que todas las columnas son variables cualitativas con excepción de la columna 'helpful' que es cuantitativa. Vamos a realizar un gráfico de barras con la cantidad de posteos por año

```python
sns.set(style="whitegrid", palette="Blues")

plt.figure(figsize=(15, 5))

ax = sns.countplot(data=user_reviews, x=user_reviews["year"], palette="Blues")

ax.set_title('Distribución de Reseñas por Año', fontsize=16)
ax.set_xlabel('Año', fontsize=14)
ax.set_ylabel('Cantidad de Reseñas', fontsize=14)

ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')

for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='baseline', fontsize=10, color='black')

sns.despine()
plt.show()
```

Podemos observar que solo hay reseñas entre los años 2010 y 2015. El año con más reseñas es el año 2014

Ahora realizaremos un gráfico con la cantidad de reseñas positivas, neutrales y negativas que hubieron

```python
sns.set(style="whitegrid", palette="Blues")

plt.figure(figsize=(15, 5))

ax = sns.countplot(data=user_reviews, x=user_reviews["sentiment_analysis"], palette="Blues")

ax.set_title('Reseñas positivas, neutrales y negativas', fontsize=16)
ax.set_xlabel('Sentiment Analysis', fontsize=14)
ax.set_ylabel('Cantidad', fontsize=14)

ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')

for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='baseline', fontsize=10, color='black')

sns.despine()
plt.show()
```

Podemos observar que las reseñas por lo general suelen ser positivas

# australian_users_items

Importamos el dataset de users items que generamos en el archivo 'ETL_for_EDA.ipynb'

```python
user_items = pd.read_csv('user_items_eda.csv')
```

```python
user_items
```

Procedemos a realizar un chequeo de duplicados, nulos y el tipo de datos.

```python
h.data_info(user_items)
```

Como podemos ver, en este caso tampoco hay duplicados ni valores faltantes ya que nos encargamos de ellos anteriormente en el archivo 'ETL_australian_users_items.ipynb' y el tipo de dato para cada columna pareciera ser el correcto.

```python
user_items.info()
```

```python
user_items
```

Podemos observar que las columnas 'playtime_forever' y 'playtime_2weeks' son variables cuantitativas. Mientras que las demás columnas son variables cualitativas

```python
top_games = user_items['item_name'].value_counts().head(10)
plt.figure(figsize=(10, 6))
top_games.sort_values().plot(kind='barh')
plt.xlabel('Cantidad')
plt.ylabel('Juegos')
plt.title('Top 10 Juegos Más Consumidos')

plt.show()
```

Como podemos ver, el producto más consumido es 'Dota 2 Test'. Ahora realizaremos un top 10 de los usuarios con mayor cantidad de items utilizando la columna 'items_count' (y eliminando los duplicados)

```python
df_unique = user_items.drop_duplicates(subset=['user_id'], keep='first')
top_users = df_unique.nlargest(10, 'items_count')
plt.figure(figsize=(10, 6))
top_users.sort_values('items_count').plot(kind='barh', x='user_id', y='items_count')
plt.xlabel('Cantidad de Items')
plt.ylabel('Usuarios')
plt.title('Top 10 Usuarios con Mayor Cantidad de Items')
plt.show()
```

Podemos ver que el usuario con más artículos consumidos es phrostb. Ahora realizamos un top 10 de los usuarios con más horas jugadas ('playtime_forever')

```python
top_users_playtime = user_items.groupby('user_id')['playtime_forever'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 6))
top_users_playtime.sort_values().plot(kind='barh')
plt.xlabel('Horas Jugadas')
plt.ylabel('Usuarios')
plt.title('Top 10 Usuarios con Más Horas Jugadas')
plt.show()
```

Podemos ver que el usuario con más horas jugadas es REBAS_AS_F-T. Ahora creamos un Stripplot y un Boxplot para visualizar la distribución de la cantidad de items por usuario

```python
sns.set(style="whitegrid", palette="Blues")

fig, axes = plt.subplots(2, 1, figsize=(14, 8), sharex=True)

sns.stripplot(data=df_unique, x='items_count', ax=axes[0], color="orange", jitter=0.2, alpha=0.7)
axes[0].set_xlabel('Cantidad de Items')
axes[0].set_ylabel('Frecuencia')
axes[0].set_title('Distribución de Cantidad de Items por Usuario')

sns.boxplot(data=df_unique, x='items_count', ax=axes[1], palette="Blues")
axes[1].set_xlabel('Cantidad de Items')
axes[1].set_ylabel('Frecuencia')
axes[1].set_title('Boxplot de Cantidad de Items por Usuario')

plt.tight_layout()
plt.show()
```

Y ahora creamos un Stripplot y un Boxplot para visualizar la distribución de la cantidad total de horas jugadas por usuario

```python
users_playtime = user_items.groupby('user_id')['playtime_forever'].sum().sort_values(ascending=False)
users_playtimedf = pd.DataFrame(users_playtime)

sns.set(style="whitegrid", palette="Blues")

fig, axes = plt.subplots(2, 1, figsize=(14, 8), sharex=True)

# Subplot 1 - Stripplot
sns.stripplot(x=users_playtime, ax=axes[0], color="orange", jitter=0.2, alpha=0.7)
axes[0].set_xlabel('Tiempo de Juego (en horas)')
axes[0].set_ylabel('Frecuencia')
axes[0].set_title('Distribución de Tiempo de Juego por Usuario')

# Subplot 2 - Boxplot
sns.boxplot(x=users_playtime, ax=axes[1], palette="Blues")
axes[1].set_xlabel('Cantidad Total de Horas Jugadas por Usuario')
axes[1].set_ylabel('Frecuencia')
axes[1].set_title('Boxplot de Cantidad Total de Horas Jugadas por Usuario')

plt.tight_layout()
plt.show()
```

# output_steam_games

Importamos el dataset steam_games que generamos en el archivo 'ETL_for_EDA.ipynb'

```python
steam_games = pd.read_csv('steam_games_eda.csv')
```

Reemplazamos los valores '&amp' por '&' en la columna 'genres'

```python
steam_games['genres'] = steam_games['genres'].str.replace('&amp;', '&')
```

```python
steam_games
```

Realizamos un chequeo de nulos, duplicados y el tipo de dato de las columnas

```python
h.data_info(steam_games)
```

Como podemos ver ya no hay nulos ni duplicados porque nos hemos encargado de ellos anteriormente y el tipo de dato para cada columna pareciera ser el correcto. Todas las variables son cualitativas a excepción de la columna 'price' que es variable cuantitativa. Ahora procedemos a graficar la cantidad de juegos que hay por género

```python
genre_counts = steam_games['genres'].value_counts()
plt.figure(figsize=(10, 6))
genre_counts.sort_values().plot(kind='barh', color='cyan')
plt.xlabel('Cantidad')
plt.ylabel('Géneros')
plt.title('Cantidad de juegos por género')

plt.show()
```

Como podemos ver, el género Indie es el predominante. Ahora veremos la cantidad de lanzamientos que hubo en cada año

```python
year_releases = steam_games["release_year"].value_counts()
```

```python
year_releases = year_releases.sort_index()

sns.set(style="whitegrid")

plt.figure(figsize=(17, 8))

ax = sns.barplot(x=year_releases.index, y=year_releases.values, palette="Blues")

for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points', fontsize=10)

ax.set(xlabel='Año de lanzamiento', ylabel='Cantidad de juegos lanzados',
       title='Cantidad de juegos por año de lanzamiento')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')

plt.grid(True, axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

Hasta 2007 no habían lanzamientos de otra desarrolladora que no fuera Valve, por eso hay una baja cantidad de lanzamientos antes de ese año. Pero luego de 2007 vemos como hay un gran crecimiento en la cantidad de lanzamientos por año. Ahora procederemos a hacer un top 10 de las desarrolladoras con más juegos desarrollados

```python
genre_counts = steam_games['developer'].value_counts().head(10)
plt.figure(figsize=(10, 6))
genre_counts.sort_values().plot(kind='barh', color='cyan')
plt.xlabel('Cantidad')
plt.ylabel('Desarrolladoras')
plt.title('Cantidad de juegos por desarrolladora')

plt.show()
```

Como podemos observar, la desarrolladora 'Ubisoft - San Francisco' es la que predomina en cantidad de juegos desarrollados
