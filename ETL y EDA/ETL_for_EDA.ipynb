# Preparación de datasets para el EDA

Haremos un ETL para preparar los datasets para el EDA

```python
import pandas as pd
import warnings
```

Filtramos las advertencias para que el código quede más prolijo

```python
warnings.filterwarnings("ignore")
```

Importamos el dataset limpio de output_steam_games

```python
steam_games = pd.read_csv('output_steam_games.csv')
```

```python
steam_games
```

Filtramos valores 'No data' en las columnas seleccionadas

```python
steam_games = steam_games[(steam_games['developer'] != 'No data')]
```

```python
steam_games = steam_games[(steam_games['release_date'] != 'No data')]
```

```python
steam_games = steam_games[(steam_games['genres'] != 'No data')]
```

```python
steam_games
```

Creamos la columna 'release_year' quedandonos con el año de 'release_date' y eliminamos los nulos que queden. También convertimos la nueva columna a entero

```python
steam_games["release_year"] = steam_games["release_date"].str[:4]
```

```python
steam_games.drop('release_date', axis=1, inplace=True)
```

```python
steam_games["release_year"] = pd.to_numeric(steam_games["release_year"], errors='coerce')
```

```python
steam_games.dropna(inplace=True)
```

```python
steam_games['release_year'] = steam_games['release_year'].astype(int)
```

```python
steam_games
```

Convertimos la columna 'genres' a listas y utilizamos explode

```python
steam_games['genres'] = steam_games['genres'].apply(eval)
```

```python
steam_games = steam_games.explode('genres')
```

```python
steam_games
```

Guardamos el Dataframe en formato CSV

```python
steam_games.to_csv('steam_games_eda.csv', index=False)
```

Ahora trabajaremos con el dataset limpio y actualizado de australian_user_reviews que obtuvimos tras realizar el procedimiento de Feature Engineering en el archivo 'Feature_Engineering.ipynb' tras agregar la columna 'sentiment_analysis'

```python
user_reviews = pd.read_csv('australian_user_reviews_actualizado.csv')
```

```python
user_reviews
```

En la columna 'posted' nos quedaremos solo con el año. Eliminamos, nulos y pasamos la columna 'year' a tipo entero

```python
user_reviews['year'] = user_reviews['posted'].str.extract(r'(\d{4})')
```

```python
user_reviews.drop('posted', axis=1, inplace=True)
```

```python
user_reviews.dropna(inplace=True)
```

```python
user_reviews['year'] = user_reviews['year'].astype(int)
```

```python
user_reviews
```

Guardamos en formato CSV

```python
user_reviews.to_csv('user_reviews_eda.csv', index=False)
```

Ahora importamos el dataset limpio de australian users items

```python
user_items = pd.read_csv('australian_users_items.csv')
```

```python
user_items
```

La columna 'playtime_forever' está en minutos, dividimos por 60 para pasarla a horas

```python
user_items['playtime_forever'] = (user_items['playtime_forever'] / 60).round(2)
```

```python
user_items
```

Guardamos el Dataframe en formato CSV

```python
user_items.to_csv('user_items_eda.csv', index=False)
```
