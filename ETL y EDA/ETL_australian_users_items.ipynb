# ETL de australian_users_items

```python
import pandas as pd
import herramientas as h
```

Importamos el archivo australian_users_items a un Dataframe

```python
with open('australian_users_items.json', 'r', encoding='utf-8') as file:
    data = [eval(line) for line in file]

df_user_items = pd.DataFrame(data)
```

```python
df_user_items.head()
```

Podemos observar que este Dataframe tiene columnas anidadas, las desanidaremos

```python
user_items_exploded = df_user_items.explode('items')
user_items_normalized = pd.json_normalize(user_items_exploded['items'])
user_items_exploded.reset_index(drop=True, inplace=True)
user_items_normalized.reset_index(drop=True, inplace=True)
user_items_final = pd.concat([user_items_exploded.drop('items', axis=1), user_items_normalized], axis=1)
```

```python
df_user_items = user_items_final
```

```python
h.data_info(df_user_items)
```

Como hemos visto, el Dataframe posee tanto valores nulos como filas duplicadas, las filas duplicadas las eliminaremos y los valores nulos representan un 0.33% en todos los casos, al ser un porcentaje muy bajo asumiremos que no aportan información relevante, por lo que también eliminaremos las filas con valores nulos.

```python
df_user_items.drop_duplicates(inplace=True)
```

```python
df_user_items.dropna(inplace=True)
```

```python
h.data_info(df_user_items)
```

```python
df_user_items.head()
```

Cambiamos el tipo de dato de las columnas "playtime_forever" y "playtime_2weeks" de flotante a entero

```python
df_user_items['playtime_forever'] = df_user_items['playtime_forever'].astype(int)
df_user_items['playtime_2weeks'] = df_user_items['playtime_2weeks'].astype(int)
```

```python
h.data_info(df_user_items)
```

Guardamos el Dataframe limpio en un archivo CSV

```python
df_user_items.to_csv('australian_users_items.csv', index=False)
```
