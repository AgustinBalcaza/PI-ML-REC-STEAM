# ETL de australian_user_reviews

```python
import ast
import pandas as pd
import herramientas as h
```

Importamos el archivo australian_user_reviews a un Dataframe

```python
archivo = 'australian_user_reviews.json'

rows = []

with open (archivo, encoding='utf-8') as f:
    for line in f.readlines():
        rows.append(ast.literal_eval(line))

df_user_reviews = pd.DataFrame(rows)
```

```python
df_user_reviews.head()
```

```python
df_user_reviews.tail()
```

Como podemos notar a primera vista, este Dataframe tiene columnas anidadas, por lo que procedemos a desanidarlas

```python
unpacked_data = []
for index, row in df_user_reviews.iterrows():
    user_id = row['user_id']
    user_url = row['user_url']
    for review_dict in row['reviews']:
        review_dict['user_id'] = user_id
        review_dict['user_url'] = user_url
        unpacked_data.append(review_dict)

df_user_reviews_desanidado = pd.DataFrame(unpacked_data)
```

```python
df_user_reviews_desanidado.head()
```

```python
df_user_reviews = df_user_reviews_desanidado
```

Guardamos el Dataframe en formato CSV para poder trabajar mejor

```python
df_user_reviews.to_csv('df_user_reviews.csv', index=False)
```
