```python
import json
import pandas as pd
import herramientas as h
import warnings
```

Empezamos por filtrar las advertencias para que de esta manera el código quede más prolijo

```python
warnings.filterwarnings("ignore")
```

Importamos el archivo output_steam_games a un Dataframe y comenzamos a analizarlo

```python
archivo = 'output_steam_games.json'

with open(archivo) as file:
    content = file.read()

json_obj = [json.loads(line) for line in content.split('\n') if line.strip()]

df_steam_games = pd.json_normalize(json_obj)
```

```python
df_steam_games = pd.read_csv('df_steam_games.csv')
```

Hemos visto que el Dataframe posee filas que son completamente nulas, procederemos a eliminarlas

```python
df_steam_games.dropna(how='all', inplace=True)
```

```python
df_steam_games.head()
```

```python
h.data_info(df_steam_games)
```

Eliminamos los pocos valores nulos que quedaron en las columnas "app_name", "reviews_url" e "id"

```python
df_steam_games = df_steam_games.dropna(subset=['app_name'])
df_steam_games = df_steam_games.dropna(subset=['reviews_url'])
df_steam_games = df_steam_games.dropna(subset=['id'])
```

```python
h.data_info(df_steam_games)
```

```python
df_steam_games.head()
```

Podemos notar también que las columnas "app_name" y "title" parecieran tener la misma información, por lo tanto procederemos a eliminar una

```python
df_steam_games.drop(columns=['title'], inplace=True)
```

```python
df_steam_games.head()
```

Podemos notar también que en la columna "price" hay valores como "Free to Play", por lo tanto procederemos a analizar y normalizar esa columna

```python
df_steam_games['price'][df_steam_games['price'].notnull()].unique()
```

Como podemos ver, exceptuando el caso de "Starting at $499.0" y "Starting at $449.00", todos los demás valores corresponderían a 0 ya que se tratan de productos gratis o de demos, por lo que procederemos a hacer las transformaciones. También cambiaremos la columna "id" a tipo entero.

```python
words = ['Free', 'Now', 'Demo', 'Install', 'party']

seen = []

for value in df_steam_games['price'].values:

    try:
        if any(word in value for word in words) and value not in seen:
            seen.append(value)
    except:
        continue

seen
```

```python
df_steam_games['price'] = df_steam_games['price'].replace(seen, '0')
df_steam_games['price'] = df_steam_games['price'].replace('Starting at $499.00', '499.0')
df_steam_games['price'] = df_steam_games['price'].replace('Starting at $449.00', '449.0')
df_steam_games['price'] = df_steam_games['price'].astype(float)
df_steam_games['id'] = df_steam_games['id'].astype(int)
```
